```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================================================
# STEP 1: DATAFICATION (Simulating the App Tracking)
# =========================================================
# We generate 1,000 random song plays to simulate tracking.

np.random.seed(42) # Ensures our random data is the same every time

# The app tracks the genre and how many minutes the user listened
genres_available = ["Pop", "Rock", "Hip-Hop", "Jazz"]
tracked_genres = np.random.choice(genres_available, size=1000)
minutes_listened = np.round(np.random.uniform(low=0.5, high=5.0, size=1000), 2)

# If they listen for more than 3 minutes, the app records it as a "Like" (True/False)
song_liked = minutes_listened > 3.0


# =========================================================
# STEP 2: BIG DATA (Storing the Asset)
# =========================================================
# We compile the tracked actions into a single pandas DataFrame.

music_database = pd.DataFrame({
    'Genre': tracked_genres,
    'Minutes': minutes_listened,
    'Liked': song_liked
})

print("\n--- A Peek at the 'Big Data' Asset ---")
print(music_database.head())


# =========================================================
# STEP 3: DATA SCIENCE (Extracting the Insight)
# =========================================================
# "Which genre should we recommend to this user next?"

# Filter the database to ONLY look at songs the user liked
liked_songs_only = music_database[music_database['Liked'] == True]

# Count how many liked songs belong to each genre (value_counts handles the sorting for us!)
genre_preferences = liked_songs_only['Genre'].value_counts()

print("\n--- Data Science Insight: Liked Songs per Genre ---")
print(genre_preferences)

# Visualize our findings for the business team
plt.figure(figsize=(8, 5))
genre_preferences.plot(kind='bar', color=['darkblue', 'royalblue', 'skyblue', 'lightblue'])
plt.title("Recommendation Insight: User's Favorite Genres")
plt.xlabel("Music Genre")
plt.ylabel("Total Liked Songs")
plt.xticks(rotation=0) # Keeps the genre names right-side up
plt.show()
```
