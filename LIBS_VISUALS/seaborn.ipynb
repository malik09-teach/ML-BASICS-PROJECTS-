```python
import seaborn as sns
import matplotlib.pyplot as plt

# Apply the default Seaborn theme
sns.set_theme(style="darkgrid")

# Load a built-in dataset (this is a Pandas DataFrame)
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.get_dataset_names()
```

```python
# Axes-level scatter plot
sns.scatterplot(
    data=tips, x="total_bill",  y="tip",  hue="smoker" # This automatically colors the dots by category!
    )

plt.title("Bill vs Tip Amount")
plt.show()
```

```python
# Create a histogram with a smooth KDE (Kernel Density Estimate) curve
sns.histplot(
    data=tips, 
    x="total_bill", 
    kde=True,       # Adds the smooth line
    color="purple"
)

plt.title("Distribution of Total Bills")
plt.show()
```

```python
# Compare the total bills paid on different days of the week
sns.boxplot(
    data=tips, 
    x="day", 
    y="total_bill", 
    palette="Set2" # Uses a pre-defined color palette
)

plt.title("Total Bills by Day of the Week")
plt.show()
```

```python
# This creates a separate scatter plot for Lunch and Dinner
sns.relplot(
    data=tips, 
    x="total_bill", 
    y="tip", 
    hue="smoker", 
    col="time"    # This splits the chart into columns based on Lunch/Dinner
)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset("penguins").dropna()
```

```python
# 1. Create the canvas and axes
fig, ax = plt.subplots(figsize=(8, 6))

# 2. Find all the unique species so we can loop through them
species_list = df['species'].unique()

# 3. Loop through each species, filter the data, and plot it
for species in species_list:
    # Filter the dataframe for only this specific species
    subset = df[df['species'] == species]
    
    # Plot this specific subset and give it a label for the legend
    ax.scatter(subset['flipper_length_mm'], subset['body_mass_g'], label=species)

# 4. Manually add all the text labels
ax.set_xlabel('Flipper Length (mm)')
ax.set_ylabel('Body Mass (g)')
ax.set_title('Penguin Size by Species')

# 5. Tell Matplotlib to build the legend based on the loop labels
ax.legend(title='Species')

plt.show()
```

```python
# Create the plot, color it by species (hue), and auto-generate labels/legend
sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', hue='species')

plt.show()
```
