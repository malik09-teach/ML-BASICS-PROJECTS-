```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python


x = np.linspace(0, 10, 100)
y = np.sin(x)





```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, color='blue', linewidth=2, linestyle='--', label='Sine Wave')
ax.set_title('Basic Line Plot')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude')
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(20, 150, 50)
# c = color array, s = size array, alpha = transparency
scatter = ax.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')
# Add a colorbar connected to the scatter map
fig.colorbar(scatter, ax=ax, label='Intensity')
ax.set_title('Advanced Scatter Plot')
plt.show()
```

```python
data = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(8, 4))
# Plot histogram with density normalizer
count, bins, ignored = ax.hist(data, bins=30, density=True, color='lightgray',
edgecolor='black')
# Overlay a mathematical density line
mu, sigma = 0, 1
best_fit_line = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins
- mu))**2))
ax.plot(bins, best_fit_line, color='red', linewidth=2)
ax.set_title('Histogram with Best Fit Line')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 6))
# Create a 2x2 grid framework
grid = fig.add_gridspec(2, 2)
# Chart 1 takes the entire top row
ax1 = fig.add_subplot(grid[0, :])
ax1.plot(np.random.randn(100).cumsum())
ax1.set_title('Main Timeline (Spans 2 columns)')
# Chart 2 takes bottom-left
ax2 = fig.add_subplot(grid[1, 0])
ax2.hist(np.random.randn(100), bins=10)
# Chart 3 takes bottom-right
ax3 = fig.add_subplot(grid[1, 1])
ax3.scatter(np.random.rand(20), np.random.rand(20))
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

correlation_matrix = np.random.rand(10, 10)
fig, ax = plt.subplots(figsize=(6, 5))

# imshow renders a 2D matrix as colored pixels
cax = ax.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')

# Add a colorbar
fig.colorbar(cax, label='Correlation Coefficient')

# Add exact text numbers inside the pixels
for i in range(10):
    for j in range(10):
        # Round to 1 decimal for readability
        val = round(correlation_matrix[i, j], 1)
        ax.text(j, i, str(val), ha='center', va='center', color='black', fontsize=8)

ax.set_title('Feature Correlation Heatmap')
plt.show()
```

```python
# Import the 3D toolkit
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 6))
# Specify that this specific axes is 3D
ax = fig.add_subplot(111, projection='3d')
# Create a meshgrid (a 2D plane of coordinates)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
# Z is a mathematical function of X and Y
Z = np.sin(np.sqrt(X**2 + Y**2))
# Plot the 3D surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_title('3D Sine Wave Surface')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Initialize the figure
fig = plt.figure(figsize=(8, 5))

# 2. Define a 2x2 grid with custom column widths
gs = fig.add_gridspec(nrows=2, ncols=2, width_ratios=[2, 1], hspace=0.3, wspace=0.3)

# 3. Slice the grid to add individual subplots (0-indexed)
ax_main = fig.add_subplot(gs[:, 0])      # Spans all rows, first column
ax_top_right = fig.add_subplot(gs[0, 1])  # Top row, second column
ax_bot_right = fig.add_subplot(gs[1, 1])  # Bottom row, second column

# 4. Add titles to visualize the placement
ax_main.set_title("Main Plot (gs[:, 0])")
ax_top_right.set_title("Top Right (gs[0, 1])")
ax_bot_right.set_title("Bottom Right (gs[1, 1])")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Initialize the figure
fig = plt.figure(figsize=(12, 6))

# 2. Define a 1x2 grid where the 3D surface gets more horizontal space
gs = fig.add_gridspec(nrows=1, ncols=2, width_ratios=[1.5, 1])

# 3. Create a 3D subplot on the left side
ax1 = fig.add_subplot(gs[0, 0], projection='3d')

# Data generation
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the 3D surface
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10, pad=0.1)
ax1.set_title('3D Sine Wave Surface')

# 4. Create a 2D subplot on the right side (e.g., a top-down contour view)
ax2 = fig.add_subplot(gs[0, 1])
contour = ax2.contourf(X, Y, Z, cmap='viridis')
fig.colorbar(contour, ax=ax2, shrink=0.5, aspect=10)
ax2.set_title('2D Contour Projection')

# Adjust layout bounding box
plt.tight_layout()
plt.show()
```
