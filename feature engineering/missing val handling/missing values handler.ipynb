```python
import pandas as pd
import seaborn as sns 
import numpy as np 
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv('Customer_Data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
# Impute missing values for MINIMUM_PAYMENTS and CREDIT_LIMIT
data[['MINIMUM_PAYMENTS', 'CREDIT_LIMIT']] = imputer.fit_transform(data[['MINIMUM_PAYMENTS', 'CREDIT_LIMIT']])
```

```python
# Verify that missing values have been handled
data.isna().sum()
```

Note: The original notebook included steps for One-Hot Encoding using ColumnTransformer. Since `Customer_Data.csv` contains only numerical features (apart from the `CUSTID` column which is an identifier), these categorical encoding steps have been removed as they do not apply to this dataset.
