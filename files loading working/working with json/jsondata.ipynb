```python
import json
import pandas as pd
import os

# 1. Define paths
input_path = r"D:\data scinece lab projects\files loading working\working with json\train.json"
output_path = r"D:\data scinece lab projects\files loading working\working with json\train_fixed.jsonl"

# 2. CONVERSION STEP (Only run this if the .jsonl doesn't exist yet)
if not os.path.exists(output_path):
    print("Converting standard JSON to JSON Lines format...")
    with open(input_path, 'r') as f_in:
        raw_data = json.load(f_in) # Load the giant list once
        
    with open(output_path, 'w') as f_out:
        for record in raw_data:
            f_out.write(json.dumps(record) + '\n')
    print("Conversion complete!")

# 3. CHUNKING STEP
print("Starting chunked ingestion...")
# We use 'lines=True' because we are reading the NEW fixed file
data_reader = pd.read_json(output_path, lines=True, chunksize=5000)

for chunk in data_reader:
    print(f"Shape: {chunk.shape}")
    
    

    
    
    
```
