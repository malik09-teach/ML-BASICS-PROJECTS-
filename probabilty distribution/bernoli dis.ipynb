```python
marks=[ 51,  92,  14,  71,  60,  20,  82,  86,  74,  74,  87,  99,  23,
         2,  21,  52,   1,  87,  29,  37,   1,  63,  59,  20,  32,  75,
        57,  21,  88,  48,  90,  58,  41,  91,  59,  79,  14,  61,  61,
        46,  61,  50,  54,  63,   2, 100,  50,   6,  20,  72]


outcomes = [1 if m >= 50 else 0 for m in marks]
p_calculated = sum(outcomes) / len(outcomes)

```

```python


plt.figure(figsize=(10, 5))

plt.hist(marks, bins=10, color='skyblue', edgecolor='black')
plt.axvline(50, color='red', linestyle='dashed', linewidth=2, label='Pass Mark (50)')
plt.title('Distribution of 50 Marks')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.legend()

plt.show
```

```python
plt.figure(figsize=(10,5))
plt.bar(['Fail (0)', 'Pass (1)'], [outcomes.count(0), outcomes.count(1)], color=['red', 'green'])
plt.title(f'Bernoulli Op (p={p_calculated:.2f})')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()

```

```python
print(f"Total Marks: {len(marks)}")
print(f"Passes: {outcomes.count(1)}")
print(f"Fails: {outcomes.count(0)}")
print(f"Calculated p: {p_calculated}")

```

## BINOMIAL

```python
def binomial_pmf(k, n, p):
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
```

```python
n = 20
k_target = 10

prob_exact = binomial_pmf(k_target, n, p_calculated)
prob_at_most = sum(binomial_pmf(i, n, p_calculated) for i in range(k_target + 1))
prob_at_least = sum(binomial_pmf(i, n, p_calculated) for i in range(k_target, n + 1))

mean_binom = n * p_calculated
var_binom = n * p_calculated * (1 - p_calculated)

print(f"Probability of exactly {k_target} successes: {prob_exact:.4f}")
print(f"Probability of at most {k_target} successes: {prob_at_most:.4f}")
print(f"Probability of at least {k_target} successes: {prob_at_least:.4f}")
print(f"Mean of Binomial (np): {mean_binom:.4f}")
print(f"Variance of Binomial np(1-p): {var_binom:.4f}")
```

```python
k_values = list(range(n + 1))
pmf_values = [binomial_pmf(k, n, p_calculated) for k in k_values]

plt.figure(figsize=(10, 5))
plt.bar(k_values, pmf_values, color='lightgreen', edgecolor='black')
plt.title('Binomial Distribution PMF')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
