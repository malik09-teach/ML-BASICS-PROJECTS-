# MEAN AND VARAINCE 

```python
# MEAN 
import numpy as np

data=[28,30,32,34,36]


x=0
n=0
for i in range (len(data)):
    x=x+data[i] 
n=i+1

mean=x/n







```

```python
# VARAINCE 

sum_sq_x=0
for j in range (len(data)):
    
    cal_x=(data[j]-mean)
    print("diff:",cal_x)
    sq_x=(cal_x*cal_x)
    print ("squre:",sq_x)
    sum_sq_x+=sq_x
    print("sum:",sum_sq_x)

var=sum_sq_x/(n-1)

```

```python
print(" mean:",mean)
print ("var:",var)
```

# BERNOILE DISTRIBUION


```python
## bernoli formula the out come may be true or false 
# mean :p=sum X / len x
# 1-p varaince
## formula  : p(X=x)=p^x (1-p)^x-1

marks = [78, 45, 90, 62, 30, 55, 88, 40, 72, 25]

x_sequence = [1 if m >= 50 else 0 for m in marks]


p_mean = sum(x_sequence) / len(x_sequence)

def calculate_bernoulli_for_sequence(seq, p):
    results = []
    
    for i, val in enumerate(seq):
    
        prob = (p ** val) * ((1 - p) ** (1 - val))
        results.append((i + 1, val, prob))
    return results

```

```python
seq_probabilities = calculate_bernoulli_for_sequence(x_sequence, p_mean)

print("Seq# | Value(x) | P(X=x)")
for item in seq_probabilities:
    print(f"{item[0]:<4} | {item[1]:<8} | {item[2]:.2f}")
```

```python
# ploting 
import matplotlib.pyplot as plt


plt.figure(figsize=(10,5))

plt.title("bernoli grap")
plt.bar(['failure (0)','success(1)'],[x_sequence.count(0),x_sequence.count(1)],color=['red','blue'])
plt.ylabel("num of students")
plt.show()



```
