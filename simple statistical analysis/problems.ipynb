# MEAN AND VARAINCE 

```python
# MEAN 
import numpy as np

data=[28,30,32,34,36]


x=0
n=0
for i in range (len(data)):
    x=x+data[i] 
n=i+1

mean=x/n







```

```python
# VARAINCE 

sum_sq_x=0
for j in range (len(data)):
    
    cal_x=(data[j]-mean)
    print("diff:",cal_x)
    sq_x=(cal_x*cal_x)
    print ("squre:",sq_x)
    sum_sq_x+=sq_x
    print("sum:",sum_sq_x)

var=sum_sq_x/(n-1)

```

```python
print(" mean:",mean)
print ("var:",var)
```
