```python
sample_50 = data.sample(n=50, random_state=42)
s50_mean = sample_50['Exam_Score'].mean()
s50_median = sample_50['Exam_Score'].median()
s50_mode = sample_50['Exam_Score'].mode()[0]
s50_var = sample_50['Exam_Score'].var(ddof=1)
```

```python
sample_50
```

```python
print ("mode=",s50_mode)
print("median=",s50_median)
print ("mean=",s50_mean)
print ("variance=",s50_var)
```

```python
scores = sample_50['Exam_Score'].tolist()
n = len(scores)

manual_mean = sum(scores) / n

sorted_scores = sorted(scores)
mid = n // 2

manual_median = (sorted_scores[mid - 1] + sorted_scores[mid]) / 2


squared_diff = [(x - manual_mean)**2 for x in scores]  

# taking the n-1 due to the delta degree of fredom is the 1
manual_var = sum(squared_diff) / (n - 1)

print(f"Manual Mean: {manual_mean}")
print(f"Manual Median: {manual_median}")
print(f"Manual Variance: {manual_var}")
```

# VISUAL REPRESNTATION

```python
## the random data and their states with new sample 

sizes = [30, 60, 120]
states = [1, 2, 3]
results = []
# instead of it you can use the zip() function 
for i in range(len(sizes)):
    size = sizes[i]
    state = states[i]
    sample = data['Exam_Score'].sample(n=size, random_state=state)
    results.append({
        'Size': size,
        'Mean': sample.mean(),
        'Median': sample.median(),
        'Variance': sample.var(ddof=1) 
    })
```

```python
comparison_data = pd.DataFrame(results)
print("\nSampling Experiment Results:\n", comparison_data)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Exam_Score'], color='blue', label='Population', kde=True, stat="density")
sns.histplot(sample_50['Exam_Score'], color='red', label='Sample (n=50)', kde=True, stat="density")
plt.title("Population vs Sample Exam Score Distribution")
plt.legend()
plt.show()
```
